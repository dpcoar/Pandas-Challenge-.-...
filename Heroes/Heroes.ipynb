```python
import pandas as pd
import numpy as np
```

```python
file_to_load = "purchase_data.csv"
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

```python
no_dup=purchase_data.drop_duplicates("SN",keep="first")
num_players = no_dup.count()[0]
pd.DataFrame({"Total Players": [num_players]})
```

```python
unique_items = len(purchase_data["Item ID"].unique())
average_price = (purchase_data["Price"].mean())
total_purchases = (purchase_data["Purchase ID"].count())
total_revenue= (purchase_data["Price"].sum())
summary=pd.DataFrame({"Number of Unique Items":[unique_items] , "Average Price":[average_price] , "Number of Purchases":[total_purchases] , "Total Revenue":[total_revenue]})
summary ["Average Price"] = summary["Average Price"].map("${:,.2f}".format)
summary ["Number of Purchases"] = summary["Number of Purchases"].map("{:,}".format)
summary ["Total Revenue"] = summary["Total Revenue"].map("${:,.2f}".format)
summary
```

```python
gender_totals = no_dup["Gender"].value_counts()
gender_percent = gender_totals/num_players
gender_demo = pd.DataFrame({"Total Count":gender_totals,"Percentage of Players":gender_percent})
gender_demo ["Percentage of Players"] = gender_demo["Percentage of Players"].map("{:,.2%}".format)
gender_demo
```

```python
bins = [0,9,14,19,24,29,34,39,100]
age_ranges = ["<10", "10-14","15-19", "20-24", "25-29", "30-34", "35-39", "40+"]


#Create a  new column for age groups and then groupby Age Groups
no_dup["Age Groups"] = pd.cut(no_dup["Age"], bins, labels=age_ranges)
age_df = no_dup.groupby(["Age Groups"])
age_count = no_dup["Age Groups"].value_counts()
age_percentage = age_count / num_players
age_demographics = pd.DataFrame({"Total Count": age_count,
                             "Percentage of Players":age_percentage})

age_demographics["Percentage of Players"] = age_demographics["Percentage of Players"].map("{:,.2%}".format) 
age_demographics = age_demographics.reindex(["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"])
age_demographics
```

```python
bins2 = [0,9,14,19,24,29,34,39,100]
age_ranges2 = ["<10", "10-14","15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
purchase_data["Age Groups"] = pd.cut(purchase_data["Age"], bins, labels=age_ranges2)
p_age_df = purchase_data.groupby(["Age Groups"])
age_purchase = purchase_data["Age Groups"].value_counts()
age_purchase

age_average_price = p_age_df["Price"].mean()
age_price = p_age_df["Price"].sum()
normalized_age = age_price / age_purchase
age_analysis = pd.DataFrame({"Purchase Count": age_purchase,
                             "Average Purchase Price":age_average_price,
                            "Total Purchase Value":age_price,
                            "Normalized Totals": normalized_age})

age_analysis["Average Purchase Price"] = age_analysis["Average Purchase Price"].map("${:,.2f}".format) 
age_analysis["Total Purchase Value"] = age_analysis["Total Purchase Value"].map("${:,.2f}".format) 
age_analysis["Normalized Totals"] = age_analysis["Normalized Totals"].map("${:,.2f}".format) 
age_analysis = age_analysis[["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Normalized Totals"]]
age_analysis = pd.concat([age_analysis.loc[["<10"],:], age_analysis.drop("<10", axis=0)], axis=0)

age_analysis
```

```python
top_sn = purchase_data.groupby(["SN"])
total_by_sn = top_sn.sum()["Price"]
avg_by_sn = top_sn.mean()["Price"]
count_sn = top_sn.count()["Price"]

top_user_df = pd.DataFrame({"Purchase Count":count_sn,
                            "Average Purchase Price":avg_by_sn,
                            "Total Purchase Price": total_by_sn
                            })

sorted_df = top_user_df.sort_values("Total Purchase Price",ascending=False)
sorted_df["Average Purchase Price"] = sorted_df["Average Purchase Price"].map("${:,.2f}".format) 
sorted_df["Total Purchase Price"] = sorted_df["Total Purchase Price"].map("${:,.2f}".format) 
sorted_df = sorted_df[["Purchase Count", "Average Purchase Price", "Total Purchase Price"]]
sorted_df.head(5)
```

```python
item_group = purchase_data.set_index(["Item ID", "Item Name"])
item_group = item_group.groupby(level=["Item ID", "Item Name"])
total_price_id = item_group.sum()["Price"]
avg_price_id = item_group.mean()["Price"]
count_id = item_group.count()["Price"]

pop_items = pd.DataFrame({ 
                         "Count":count_id,
                            "Average Purchase Price":avg_price_id,
                            "Total Purchase Price": total_price_id,
                            })


pop_sorted = pop_items.sort_values("Count",ascending=False)
pop_sorted["Average Purchase Price"] = pop_sorted["Average Purchase Price"].map("${:,.2f}".format) 
pop_sorted["Total Purchase Price"] = pop_sorted["Total Purchase Price"].map("${:,.2f}".format) 
pop_sorted = pop_sorted[["Count", "Average Purchase Price", "Total Purchase Price"]]
pop_sorted.head(5)
```
